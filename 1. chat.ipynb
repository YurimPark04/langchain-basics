```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="deepseek-r1:1.5b")  # 사전에 ollama에서 pull을 받아와야함
llm.invoke("What is the capital of France?") 
# 추론 모델 -> AIMessage의 content가 LLM의 응답
# 추론 모델이라서 <think> 이런 요소들이 들어가있을 수 있다
```

```python
llm2 = ChatOllama(model="llama3.2:1b")
llm2.invoke("What is the capital of France?") 
```
